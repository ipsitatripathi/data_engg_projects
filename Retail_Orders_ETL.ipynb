```python
#remove not available, unknown ship mode data 
df = pd.read_csv(r'C:\Users\ADMINUSER\orders.csv', na_values=['Not Available', 'unknown'])
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
#rename columns and replace space with underscore

df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(" ", "_")
df.head()
```

```python
#add new columns discount, sale price and profit

df['discount']=df['list_price'] * df['discount_percent'] * 0.01
df['sale_price'] = df['list_price'] - df['discount']
df['profit'] = df['sale_price'] - df['cost_price']
df.head()
```

```python
df.dtypes
```

```python
#convert order date to datetime

df['order_date'] = pd.to_datetime(df['order_date'], format="%Y-%m-%d")
df.dtypes
```

```python
#drop list price, cost price and discount percent

df.drop(columns=['list_price','cost_price','discount_percent'], inplace=True)
df.head()
```

```python
#load the data into sql server

import sqlalchemy as sal
engine = sal.create_engine('mssql://DESKTOP-T0IMGJF\SQLEXPRESS/master?driver=ODBC+DRIVER+17+FOR+SQL+SERVER')
conn = engine.connect()
df.to_sql('df_orders', con=conn, index=False, if_exists='replace')
```
